```python
dates = []
with open('mbox.txt', 'r') as fh:
    for line in fh.readlines():
        if line.startswith('Date: ') and line.endswith(')\n'):
            dates.append(line)
        
```

```python
hours = {}
for date in dates:
    hour = date.split()[2].split(':')[0]
    hours[hour] = hours.get(hour, 0) + 1
    
for hour in sorted(hours):
    print('В %s было отправлено %s писем' % (hour, hours[hour]))
```

```python
my_table = soup.find_all('table')[0]
my_table
```

```python
len(soup.find_all('table', {'border':1}))
```

```python
rows = my_table.find_all('tr')
rows
```

```python
surnames = []
names = []
ages = []

```

```python
surnames.append(rows[0].find_all('th')[0].text)
for row in rows[1:]:
    surnames.append(row.find_all('td')[0].text)
    
print(surnames)
```

```python
def get_column(rows, col_number):
    lst = []
    lst.append(rows[0].find_all('th')[col_number].text)
    
    for row in rows[1:]:
        lst.append(row.find_all('td')[col_number].text)
        
    return lst
```

```python
get_column(rows, 2)
```

```python
surnames = get_column(rows, 0)
names = get_column(rows, 1)
ages = get_column(rows, 2)
```

```python
with open('my_table.txt', 'w', encoding='utf8') as fh:
    for idx in range(len(surnames)):
        fh.write("%s, %s, %s\n" % (surnames[idx], names[idx], ages[idx]))

```
