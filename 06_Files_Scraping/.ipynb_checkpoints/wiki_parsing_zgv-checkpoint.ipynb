```python
import requests
from bs4 import BeautifulSoup
```

```python
link = 'https://en.wikipedia.org/wiki/List_of_nuclear_weapons_tests'
soup = BeautifulSoup(requests.get(link).text, 'lxml')

print(soup.prettify())
```

```python
len(soup.find_all('table', {'class':'wikitable sortable'}))
```

```python
my_table = soup.find_all('table', {'class':'wikitable sortable'})[1]
print(my_table.prettify())
```

```python
def get_column(rows, col_number):
    lst = []
    lst.append(rows[0].find_all('th')[col_number].text)

    for row in rows[1:]:
        lst.append(row.find_all('td')[col_number].text)
        
    return lst
```

```python
rows = my_table.find_all('tr')
get_column(rows, 0)
```

```python
dates = get_column(rows, 0)
yields = get_column(rows, 1)
countries = get_column(rows, 3)
print(dates)
print(yields)
print(countries)
```

```python
my_table_dict = {}
def get_column_dict(rows, col_number):
    lst = []

    for row in rows[1:]:
        lst.append(row.find_all('td')[col_number].text)
        
    my_table_dict[rows[0].find_all('th')[col_number].text.strip()] = lst
    
get_column_dict(rows, 0)
get_column_dict(rows, 1)
get_column_dict(rows, 3)
#print(my_table_dict)
print(my_table_dict.keys())
my_table_dict['Country'][:10]
```

## Задания

Напишите функцию, которая берет аргументом название страны и возвращает (return) среднюю мощность взрыва для этой страны (нужно сложить все значения из колонки 'Yield (megatons)', которым соответствует страна, например, США, и раделить на количество этих значений). Для подсчета используйте списки, которые мы извлекли выше.

```python
def get_avg_yield(country, countries, yields):
    country_yields = [float(yields[i]) for i in range(len(countries)) if countries[i].lower() == country.lower()]
    return sum(country_yields) / len(country_yields)

get_avg_yield('Soviet Union', countries, yields)
```

Из списка Country оставьте только уникальные значения для стран и запустите вашу функцию в цикле для каждого значения Country. Внутри цикла сделайте следующий вывод "{название страны}: средняя мощность взрыва {средняя мощность} мегатон"


```python
for country in set(countries[1:]):
    print("%s: %.2f мегатонн" % (country, get_avg_yield(country, countries, yields)))
```

Создайте словарь, ключом в котором будет страна, а значением список из количества испытаний и максимальной мощности взрыва.

```python
def get_test_data(country, countries, yields):
    country_yields = [float(yields[i]) for i in range(len(countries)) if countries[i].lower() == country.lower()]
    return [max(country_yields), len(country_yields)]

country_tests = {}
for country in set(countries[1:]):
    country_tests[country] = get_test_data(country, countries, yields)
    
country_tests
```
