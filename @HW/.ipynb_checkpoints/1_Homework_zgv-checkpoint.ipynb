#### Задание 1. Проверка СНИЛСа (0.25 балла)
Даны три строки, содержащие номера СНИЛС. Правильно записанный СНИЛС содержит 11 символов и ровно два дефиса (после первой и второй групп цифр). Проверьте три СНИЛСа на правильность и для каждого выведите True/False в зависимости от того правильная запись или нет.

**NB** Задание сформулировано некорректно. В СНИЛС может быть как 2 дефиса, так и 3. 
![](https://upload.wikimedia.org/wikipedia/commons/c/cd/%D0%A1%D0%9D%D0%98%D0%9B%D0%A1.jpg)

В обоих вариантах группы цифр имеют по 3 символа, а контрольная сумма - 2.

Я реализовал проверку в соответствии с реальной структурой СНИЛС

```python
snils = ['444-555-344-09', '134-456-872 77','134 456-872-77']
# ваш код здесь

digits = '0123456789'

def check_snils(s):
    
    # Фильтруем неправильные символы
    s = ''.join([c for c in s if c in digits + '- '])
    
    if len(s) != 14: # неправильная длина
        return False
        
    #приводим к одному формату, с тремя дефисами
    if s[-3]  == ' ':
        s = s[:-3] + '-' + s[-3:]
        
    components = s.split('-') # разделение на группы символов
    
    if len(components) != 4: # неправильное число компонентов
        return False
        
    if min(map(len, components[:-1])) != 3: # Группы символов, кроме последней, имеют длину не 3
        return False
    
    return True

print(list(map(check_snils, snils)))

for s in snils:
    print('Результат проверки СНИЛС %s: %s' % (s, check_snils(s)))

    
```

#### Задание 2. Проверка номера телефона (0.25 балла)
Даны два номера телефона. Правильный номер телефона начинается с '+7' и содержит 10 цифр (не считая '+7') Проверьте правильность номеров и выведите True, в случае, если он правильный (False если нет).

```python
phone = ['+89554362828', '+795543628283']
# ваш код ниже

def check_phone(s):

    if s[:2] != '+7': # неправильный код страны
        return False
    
    if len(s[2:]) != 10: # неправильная длина номера
        return False
    
    if (not min(d in digits for d in s[2:])): # Есть символы кроме цифр
        return False
    
    return True

list(map(check_phone, phone))
```

```python
phone2 = ['+89554362828', '+795543628283', '+71234567890', '+7123456789a', '+71234567890']

{s : check_phone(s) for s in phone2}
```

```python
marks = {'Mary' : [5, 8, 9, 10, 3, 5, 6, 6],
         'John' : [3, 3, 6, 8, 2, 1, 8, 5],
         'Alex' : [4, 4, 7, 4, 7, 3, 2, 9],
         'Patricia' : [2, 1, 6, 8, 2, 3, 7, 4]}

categories = {'отлично' : [8, 9, 10],
              'хорошо' : [6, 7],
              'удовлетворительно' : [4, 5],
              'неуд' : [0, 1, 2, 3]}

# Ваш код здесь
grade = int(input('Введите курс (1-8): '))

grade_marks = [marks[s][grade - 1] for s in marks]
avg_mark = round(sum(grade_marks)/len(grade_marks))

print('Курс %s - %s' % (grade, avg_mark))
```

#### Задание 8. Магнолия (2 балла)

В магазине "Магнолия" проходит акция - даются скидки на товары, название которых начинается на заданную букву.

У Маши есть список покупок, состоящий из строк вида "цена, название, количество". Помогите ей посчитать, сколько денег она сэкономит (общую сумму скидки).

Количество записей в Машином списке покупок может быть любым. Каждая запись вводится с новой строки. Записи вводятся до тех пор, пока пользователь не введет "Конец списка". После этого одной строкой вводится буква, по которой сегодня определяется скидка. Если название продукта начинается на заданную букву, то на этот продукт действует скидка 15 процентов.

Образец вывода смотрите в примере. Если ни на один товар скидка не действует, то программа должна вывести "Скидка: 0 рублей"

*Вы считали категории в список списков, где каждая строка выглядит как в образце  - 1 балл*  
`[цена (int), название (str), количество (int)]`  
Например, `[120, мандарины (1 кг), 2]`

*Подсчет скидки - 1 балл.*

**Пример ввода**  
120 руб., мандарины (1 кг), 2 кг.  
70 руб., молоко, 3 пакета  
15 руб., батон сдобный, 1 штука  
345 руб., белое вино, 1 бутылка  
120 руб., моцарелла, 2 упаковки  
Конец списка  
м  

**Вывод**

Cкидка: 103.5 руб.


```python
# Ввод списка

rows = []

print("Введите покупку (для завершения введите 'Конец списка'):")

while True:
    row = input("")
    if row.lower() == 'конец списка':
        break
    rows.append(row.split(', '))

discount_letter = input("Введите букву для получения скидки: ")

discount = 0

for r in rows:    
    price = int(r[0].split(' ')[0]) # цена
    
    name = r[1] # товар
    quantity = int(r[-1].split(' ')[0]) # количество
    if name.startswith(discount_letter):
        discount += price * quantity * 0.15
        
print('Скидка: %.2f руб' % discount)
```

###### Шаг 4. Первая догадка (1 балл)
Давайте спросим у пользователя, где по его мнению находится корабль. Считайте из консоли догадку пользователя для ряда и колонки на сетке 5x5. Сохраните результат в переменные guess_row и guess_col. Если пользователь вводит число не из интервала - выведите сообщение об ошибке ("Числа должны быть в интервале от 1 до 5") и попросить ввести число заново. Это должно быть реализовано в цикле while (принимаем на ввод значения до тех пор, пока пользователь не введет допустимые).

```python
# ваш код здесь
def get_index(prompt, min=1, max=5):
    while True:
        ans = input('Введите %s (%s..%s): ' % (prompt, min, max))
        if len(ans) == 0: # пустой ввод, нажали Enter
            continue
        ans = int(ans)
        if ans >= min and ans <= max:
            return ans
        
def get_guess():
    guess_row = get_index('строку')
    guess_col = get_index('столбец')
    return (guess_row, guess_col)

print(*get_guess())
```

Если пользователь не угадал, нужно заменить O в сетке на X и вывести, что выстрел пришелся мимо.
