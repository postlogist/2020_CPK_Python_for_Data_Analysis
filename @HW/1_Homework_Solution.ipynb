Образец вывода смотрите в примере. Если ни на один товар скидка не действует, то программа должна вывести "Скидка: 0 рублей"

*Вы считали категории в список списков, где каждая строка выглядит как в образце  - 1 балл*  
`[цена (int), название (str), количество (int)]`  
Например, `[120, мандарины (1 кг), 2]`

*Подсчет скидки - 1 балл.*

120 руб., мандарины (1 кг), 2 кг.

345 руб., белое вино, 1 бутылка

м  

**Вывод**

```python
# ваше решение здесь
item = input()
shopping_list = []

while item != 'Конец списка':
    temp = []
    temp.append(int(item.split(', ')[0].split()[0]))
    temp.append(item.split(', ')[1])
    temp.append(item.split(', ')[2].split()[0])
    shopping_list.append(temp)
    item = input()
    
letter = input()
print(shopping_list, letter)
```

```python
discount = 0
for item in shopping_list:
    if item[1].startswith(letter):
        discount += item[0]*int(item[2])*0.15
        
print(f'Скидка: {discount} руб.')
```

```python
from random import randint

# ваш код здесь

def print_board():
    for row in board:
        print(*row)
        
trials = 1

board = [['O' for i in range(5)] for i in range(5)]

ship_row = randint(0,4)
ship_col = randint(0,4)
print(ship_row, ship_col) # выводим для отладки

while trials <= 5:
    print(f'Попытка {trials}')
    
    while True:
        guess_row, guess_col = map(int, input('Введите через пробел ряд и колонку в интервале от 1 до 5: ').split())
        if guess_row not in range(1,6) or guess_col not in range(1,6):
            print("Числа должны быть в интервале от 1 до 5")
        else:
            break
            
    if guess_row-1 != ship_row or guess_col-1 != ship_col:
        if trials != 5:
            board[guess_row-1][guess_col-1] = 'X'
            print_board()
            print(f'Попытка {trials}. Ваша догадка была {guess_row}-{guess_col}. Вы промахнулись :(')
        else:
            board[guess_row-1][guess_col-1] = 'X'
            board[ship_row][ship_col] = 'S'
            print_board()
            print(f'Попытка {trials}. Ваша догадка была {guess_row}-{guess_col}. Вы промахнулись :( \n \
            Это была последняя попытка и вы проиграли. Корабль находился на клетке {ship_row}-{ship_col}.')
    else:
        board[guess_row-1][guess_col-1] = 'W'
        print_board()
        print(f'Ваша догадка была {guess_row}-{guess_col}. Вы потопили корабль!')
        break
        
    trials += 1
```
