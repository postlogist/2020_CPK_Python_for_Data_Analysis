Для графиков с большим количеством точек, как я понял, в Pandas/Matplotlib отсутствует встроенная функция 'Jitter'. Схожего визуального (но не технического) эффекта можно добиться оперируя размерами точек и прозрачностью.

```python
def rand_jitter(arr):
    stdev = max(arr)/900
    return arr + np.random.randn(len(arr)) * stdev
table = di['table'].values.tolist()
price = di['price'].values.tolist()
plt.figure(figsize=(10,8))
plt.scatter(x = rand_jitter(table), y = rand_jitter(price), alpha = 0.002)
plt.xlim(50, 70)
```
